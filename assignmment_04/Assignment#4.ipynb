## Step 5: Extract the "Total Population" column from the data_list and assign them to a list named "pop_list"

```
pop_list= [i.split('\t', 2)[1] for i in data_list] #taking the population info from list
print(pop_list)
```

## Step 6: Extract the "Census Cost" column from the data_list and assign them to a list named "cost_list"

```
cost_list= [i.split('\t', 3)[2] for i in data_list] # taking cost values from data_list
print(cost_list)
```

```
cost = []
for c in cost_list:
    new_cost = c.replace("$", "") #taking money sign out
    cost.append(new_cost)
```

```
cost_withoutcommas = []
for e in cost:
    new_cost2 = e.replace(",", "")# taking commas out
    cost_withoutcommas.append(new_cost2)
```

```
 cst= []
for e in cost_withoutcommas:
    new_cost3 = e.replace("4.5 Billion", "4500000000") # covert the billions to numbers
    cst.append(new_cost3)
```

## Step 7: Extract the "Average Cost per Person" column from the data_list and assign them to a list named "avg_list"

```
avg_list= [i.split('\t', 4)[3] for i in data_list] #taking the avg_list from data_list
print(avg_list)
```

```
avg1 = []
for c in avg_list:
    new_avg = c.replace("cents\n","") # taking cents out
    avg1.append(new_avg)
```

```
avg1
```

```
index2=avg1.index('71.34 ') #determine the index to do the division to convert to dollars
```

```
index2
```

```
new=[]
for i in avg1:
    change = i.replace(' ','') # taking out the extra space
    new.append(change)
```

```
new
```

```
avg2 = []
for c in new:
    new_avg2 = c.replace("\n","") # taking out the \n
    avg2.append(new_avg2)
```

```
avg2
```

```
avg3 = []
for c in avg2:
    new_avg3 = c.replace("$","") # taking out the dollar sign
    avg3.append(new_avg3)
```

```
avg3
```

```
avg4 = [float(elem) for elem in avg3] # convert to float


print(avg4)
```

```
dollar_amount=avg4[18:23]
```

```
dollar_amount
```

```
cents_indollars=[]
for x in avg4[:18]:
    y = (x/100) # dividing by 100 to get the dollar amount
    cents_indollars.append(y)   
print(cents_indollars)
```

## Step 8: Coalesce the cleansed data and save them to a text file named "census_cost.csv"

```
top2_list
```
